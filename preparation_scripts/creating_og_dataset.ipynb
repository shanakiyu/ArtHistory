# CREER LE DATASET D'HISTOIRE DE L'ART


```python
import pandas as pd

df1 = pd.read_csv("BYU.csv", sep=";", engine="python", on_bad_lines="skip")
df1_art = df1[df1['Department'].str.contains("Comparative Arts and Letters", case=False, na=False)]
print(df1_art.shape)
df1_art.head()
df2 = pd.read_csv("rice.csv", sep=";", engine="python", on_bad_lines="skip")
df2_art = df2[df2['departement'].str.contains("Art History", case=False, na=False)]
print(df2_art.shape)
df2_art.head()
df3 = pd.read_csv("harvard.csv", sep=";", engine="python", on_bad_lines="skip", encoding="ISO-8859-1")
df3_art = df3[df3['department'].str.contains("Art History", case=False, na=False)]
print(df3_art.shape)
df3_art.head()
df4 = pd.read_csv("stanford.csv", sep=",", engine="python", on_bad_lines="skip")
df4_art = df4[df4['department'].str.contains("art", case=False, na=False)]
print(df4_art.shape)
df4_art.head()
df5 = pd.read_csv("Williams.csv", sep=";", engine="python", on_bad_lines="skip")
df5_art = df5[df5['department'].str.contains("Art", case=False, na=False)]
print(df5_art.shape)
df5_art.head()
df6 = pd.read_csv("Michigan-data-dpt.csv", sep=",", engine="python", on_bad_lines="skip", encoding="latin1")


df6.columns = df6.columns.str.replace('"', '').str.strip()
df6_art = df6[df6['departement'].str.contains("history of art", case=False, na=False)]

print(df6_art.shape)
df6_art.head()
```

```python
def clean_columns(df):
    df.columns = df.columns.str.strip().str.lower().str.replace('"', '').str.replace(' ', '_')
    return df

df1_art = clean_columns(df1_art)
df2_art = clean_columns(df2_art)
df3_art = clean_columns(df3_art)
df4_art = clean_columns(df4_art)
df5_art = clean_columns(df5_art)
df6_art = clean_columns(df6_art)
```

```python
df_all_art = pd.concat(
    [df1_art, df2_art, df3_art, df4_art, df5_art, df6_art],
    ignore_index=True,
    sort=False 
)
print(df_all_art.shape)
```

```python
df_all_art['name'] = df_all_art['name'].combine_first(df_all_art['nom'])
df_all_art['department'] = df_all_art['department'].combine_first(df_all_art['departement'])
df_all_art['profile_url'] = (
    df_all_art['profile_url']
    .combine_first(df_all_art['url_profil'])
    .combine_first(df_all_art['personal_page_link'])
    .combine_first(df_all_art['url'])
)
df_all_art['cv'] = (
    df_all_art['cv_url'].combine_first(df_all_art['cv_link']))
df_all_art['expertise'] = (
    df_all_art['research'].combine_first(df_all_art['research_areas'])
    .combine_first(df_all_art['academic_interests'])
    .combine_first(df_all_art['subfields'])
    .combine_first(df_all_art['expertise'])
    .combine_first(df_all_art['fields']))
df_all_art['biography_summary'] = (
    df_all_art['biography'].combine_first(df_all_art['summary'])
    .combine_first(df_all_art['presentation'])
    .combine_first(df_all_art['research']))

df_all_art['job_position'] = (
    df_all_art['profession'].combine_first(df_all_art['job'])
    .combine_first(df_all_art['position_title'])
    .combine_first(df_all_art['title'])
    .combine_first(df_all_art['status'])
)
df_all_art['contacts'] = (
    df_all_art['contact'].combine_first(df_all_art['email_address'])
    .combine_first(df_all_art['phone_number'])
    .combine_first(df_all_art['address'])
)
df_all_art['awards'] = (
    df_all_art['honor_awards'].combine_first(df_all_art['awards'])
)
df_all_art['education'] = (
    df_all_art['education'].combine_first(df_all_art['degree'])
)
df_all_art['interests'] = (
    df_all_art['interests'].combine_first(df_all_art['research_areas'])
)
df_all_art['university'] = (
    df_all_art['university'].combine_first(df_all_art['c'])
)

columns_to_drop = [
    'cv_url','cv_link', 'departement', 'nom', 'url_profil', 'personal_page_link', 'url', 'c', 'research', 'research_areas','academic_interests','subfields','expertise','fields','research_methods',
    'publications','courses_taught',
    'biography','summary','presentation','presentation_length',
    'profession','job','position_title','title','status',
    'contact','email_address','phone_number','address',
    'honor_awards','awards',
    'websites','external_stanford','profile_url', "unnamed:_11", "unnamed:_12", "unnamed:_13", "unnamed:_14", "unnamed:_15", "unnamed:_16",
    'vietnam','indochine', 'personal_website_link', "unnamed:_0", "committees", "degree", 
]

df_all_art_cleaned = df_all_art.drop(columns=columns_to_drop)




df_all_art_cleaned.to_csv("ArtHistory2.csv", index=False)
```
