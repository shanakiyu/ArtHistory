```python
def parse_tags(cell):
    if pd.isna(cell):
        return []

    if not isinstance(cell, str):
        return []

    # séparation par virgule
    tags = cell.split(",")

    # nettoyage léger
    tags = [t.strip() for t in tags if t.strip()]

    return tags
```

```python
!pip -q install openai
import json
from openai import OpenAI

client = OpenAI()

def llm_map_micro_to_macro(batch, macro_tags):
    system_msg = {
        "role": "system",
        "content": (
            "You are an expert taxonomist in art history. "
            "Map each micro-tag to exactly ONE macro-tag from the provided list. "
            "Never invent new macro-tags."
        )
    }
    user_msg = {
        "role": "user",
        "content": f"""
MACRO TAGS (choose ONLY from this list):
{json.dumps(macro_tags, ensure_ascii=False)}

MICRO TAGS:
{json.dumps(batch, ensure_ascii=False)}

RULES:
- Output JSON only.
- Keys must be EXACT micro-tags (unchanged).
- Values must be EXACT macro-tags from the list.
- If unclear: choose the closest.
- If nothing fits: "Other / Uncategorized".

Return:
{{"mapping": {{"micro": "macro"}}}}
"""
    }

    resp = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[system_msg, user_msg],
        response_format={"type": "json_object"},
        temperature=0
    )
    data = json.loads(resp.choices[0].message.content)
    return data.get("mapping", {})
```

```python
import time

BATCH_SIZE = 50
mapping = {}

for start in range(0, len(microtags), BATCH_SIZE):
    batch = microtags[start:start+BATCH_SIZE]
    print(f"{start}/{len(microtags)}")

    try:
        m = llm_map_micro_to_macro(batch, MACRO_TAGS)
        mapping.update(m)
    except Exception as e:
        print("Erreur batch:", e)
        for t in batch:
            mapping[t] = "Other / Uncategorized"

    time.sleep(0.2)

mapping_df = pd.DataFrame([{"micro_tag": k, "macro_tag": v} for k, v in mapping.items()])
mapping_df.to_csv("micro_to_macro_mapping.csv", index=False)
print("Saved micro_to_macro_mapping.csv")
files.download("micro_to_macro_mapping.csv")
```

```python
def parse_tags_cell(x):
    if not isinstance(x, str) or not x.strip():
        return []
    return [t.strip() for t in x.split(",") if t.strip()]

def micro_to_macro_list(cell):
    micro = parse_tags_cell(cell)
    macros = []
    for t in micro:
        macros.append(mapping.get(t, "Other / Uncategorized"))
    # dédoublonne en gardant l'ordre
    seen = set()
    out = []
    for m in macros:
        if m not in seen:
            seen.add(m)
            out.append(m)
    return ", ".join(out) if out else np.nan


```

```python
TECH_SCIENCE_KEYWORDS = [
    "technology",
    "science",
    "scientific",
    "health",
    "medical",
    "medicine",
    "biomedical",
    "digital humanities",
    "data",
    "computation",
    "computational",
    "ai",
    "artificial intelligence",
    "cyber",
    "virtual",
    "software",
    "biology",
    "neuroscience",
    "epidemiology",
    "public health",
    "immunology"
]
```

```python
def contains_tech_science(macro_tags, keywords):
    if pd.isna(macro_tags):
        return False
    s = str(macro_tags).lower()
    return any(k in s for k in keywords)
```

```python
df_tech_science = df[df["macro_tags"].apply(
    lambda x: contains_tech_science(x, TECH_SCIENCE_KEYWORDS)
)]

```
